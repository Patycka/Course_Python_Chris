```python
import pandas as pd
from matplotlib import pyplot as plt
```

```python
deniro_csv = "deniro.csv"

deniro_data = pd.read_csv(deniro_csv, sep=",", quotechar='"', skipinitialspace=True)
print(deniro_data)
```

```python
deniro_sorted_by_score = deniro_data.sort_values(["Score"], ascending=False)
deniro_sorted_by_score
```

```python
deniro_grouped_by_year = deniro_sorted_by_score.groupby("Year")
deniro_grouped_by_year = deniro_grouped_by_year.count()
deniro_grouped_by_year
```

```python
deniro_grouped_by_year.sort_values(["Score"], ascending=False)
```
